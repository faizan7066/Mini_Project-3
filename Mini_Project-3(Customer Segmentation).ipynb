## EDA

```python
# checking for Missing/Null values

df.isnull().sum()
```

* There are 24 missing values in Income feature

```python
# Filling null vlaues with mean

df['Income'] = df['Income'].fillna(df['Income'].mean())
```

```python
df.isnull().sum()
```

* There are no null values in data

```python
# Checking for cloumns

df.columns
```

```python
# Adding dummies to single column

df['AcceptedCmp'] = df['AcceptedCmp3'] + df['AcceptedCmp4'] + df['AcceptedCmp5'] + df['AcceptedCmp1'] + df['AcceptedCmp2']
```

```python
# Assigning real values of the numbers

df['AcceptedCmp'] = df['AcceptedCmp'].replace({0 : '1', 1 : '2', 2 : '3', 3 : '4', 4 : '5'})
```

```python
# Dropping Dummy Columns

df = df.drop(['AcceptedCmp3', 'AcceptedCmp2', 'AcceptedCmp1', 'AcceptedCmp4', 'AcceptedCmp5'], axis = 1)
```

```python
df.head()
```

```python
# Converting Dt_custmer column to datetime and Extracting Day, month and Year from it

df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'])
df['day'] = df['Dt_Customer'].apply(lambda x : x.day)
df['month'] = df['Dt_Customer'].apply(lambda x : x.month)
df['year'] = df['Dt_Customer'].apply(lambda x : x.year)
```

```python
# creating a Age column out of year of birth

df['Age'] = 2023 - df['Year_Birth']
```

```python
# Creating a column as 'children' to combine Kidhome and Teenhome

df["Children"]=df["Kidhome"]+df["Teenhome"]
```

```python
# Dropping the columns that are of no use

df = df.drop(['ID','Dt_Customer', 'Z_CostContact', 'Z_Revenue','Year_Birth','Kidhome','Teenhome'], axis = 1)
```

```python
# Scaling the feature

scaled_features = df.copy()
```

```python
# Standardizing data
from sklearn.preprocessing import StandardScaler

col_names = ['Income', 'day','month','year','Recency','MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts','MntSweetProducts', 'MntGoldProds', 'NumDealsPurchases','NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases','NumWebVisitsMonth', 'Complain', 'Response', 'Age', 'Children','AcceptedCmp']
features = scaled_features[col_names]
scaler = StandardScaler().fit(features.values)
features = scaler.transform(features.values)
```

```python
# Scaled feature

scaled_features[col_names] = features
scaled_features.head()
```

## PCA

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca.fit(X)
X= pca.transform(X)
```

```python
plt.figure(figsize=(15,6))  
plt.scatter(X[:,0],X[:,1])
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

## DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
```

```python
db=DBSCAN(eps=1.298,min_samples=256)
```

```python
y=db.fit_predict(X)
```

```python
sil=silhouette_score(X,y)
print("Silhouette Score:"+str(sil))
```

```python
plt.figure(figsize=(15,6))  
plt.scatter(X[:,0],X[:,1],c=db.labels_)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```
